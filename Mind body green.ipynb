```python
#2b. Which product/channel combination has the lowest retention rate?

##January Cohort
df20= pandasql.sqldf('''
                    select 
                       "Product Name" as product,
                      "Attribution Channel" as channel,
                       
                     sum( "Total Units") as month1_retention
                       from df
                       where 
                       "Subscription Date" between '2020-01-01' and '2020-01-31'
                       and "Month Number" = 1
                       and "Order Number" != 'skipped'
                       group by 1,2
                        ;
     ''', globals())
df20
```

```python
df21 = pandasql.sqldf('''
                      
                      select 
                         product,
                        channel,
                       sum("Total Units") as month2_rentention 
                       from 
                       ( select 
                      "Customer ID", 
                      "Attribution Channel" as channel,
                      "Purchase Date",
                       "Product Name" as product,
                      "Month Number" as month,
                      "Total Units"
                       from df
                       where 
                       "Subscription Date" between '2020-01-01' and '2020-01-31'
                       and "Month Number" >1
                       and "Order Number" != 'skipped'
                       group by 1,2,3,4,5) a
                       where  month = 2
                       group by 1,2          
                        ;
     ''', globals())
df21
```

```python
##join 
df22= pandasql.sqldf('''
                    select 
                       df20.product, 
                       df20.channel,
                       month1_retention,
                        month2_rentention

                       from 
                       df20
                       left join df21 
                       on df21.product = df20.product
                       and df21.channel = df20.channel
                        ;
     ''', globals())
df22
```

```python
df22['retention_rate'] = df22['month2_rentention']/df22['month1_retention']
df22
```
